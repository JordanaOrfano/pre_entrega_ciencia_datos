# Etapa 2: Preprocesamiento y Limpieza de Datos

### 4锔 Limpieza de datos

```python
# COPIAS PARA TRABAJAR SIN MODIFICAR LOS ORIGINALES

clientes_limpio = df_clientes.copy()
ventas_limpio = df_ventas.copy()
marketing_limpio = df_marketing.copy()
```

#### Eliminaci贸n de duplicados

```python
def eliminar_duplicados(df, nombre):
    print(f"\n--------------------------- {nombre} ---------------------------")
    filas_antes = len(df)
    dups = df.duplicated().sum()
    print(f"Filas antes: {filas_antes}")

    df_limpio = df.drop_duplicates()
    filas_despues = len(df_limpio)
    eliminadas = filas_antes - filas_despues

    print(f"Filas duplicadas eliminadas: {eliminadas}")
    return df_limpio

df_ventas = eliminar_duplicados(ventas_limpio, "Ventas")
df_clientes = eliminar_duplicados(clientes_limpio, "Clientes")
df_marketing= eliminar_duplicados(marketing_limpio,"Marketing")
```

#### Limpieza de texto

```python
def normalizacion_texto(df, nombre_df):
    print(f"\n--------------------------- {nombre_df} ---------------------------")
    columnas_texto = [c for c in df.columns if df[c].dtype == 'object']
    print(f"Columnas tipo texto detectadas: {columnas_texto if columnas_texto else 'ninguna'}")

    for columna in columnas_texto:
        # Si el nombre de la columna contiene la palabra 'fecha', se saltea
        if "fecha" in columna.lower():
            print(f"'{columna}' contiene 'fecha', se omite del proceso.")
            continue

        original = df[columna].astype(str)

        # strip
        paso_strip = original.str.strip()
        cnt_strip = (original != paso_strip).sum()

        # eliminar caracteres invisibles
        paso_invis = paso_strip.str.replace(r'[\u200b\t\r\n]', '', regex=True)
        cnt_invis = (paso_strip != paso_invis).sum()

        # eliminar s铆mbolos visibles
        paso_final = paso_invis.str.replace(r'[$,%+\-"/\']', '', regex=True)
        cnt_replace = (paso_invis != paso_final).sum()

        # actualizar el DataFrame
        df[columna] = paso_final

        print(f"'{columna}': ajustadas con strip = {cnt_strip} | car. invisibles = {cnt_invis} | ajustadas con replace = {cnt_replace}")

    return df


# Aplicaci贸n a los tres conjuntos
ventas = normalizacion_texto(ventas_limpio, "Ventas")
clientes = normalizacion_texto(clientes_limpio, "Clientes")
marketing = normalizacion_texto(marketing_limpio, "Marketing")
```

#### Corregir tipos de datos

```python
def convertir_fechas(df, columnas_fecha, nombre_df, dayfirst=True):
    if not columnas_fecha:
        print("No se indicaron columnas de fecha.")
        return df

    for col in columnas_fecha:
        if col not in df.columns:
            print(f"(omitido) La columna '{col}' no existe en {nombre_df}.")
            continue

        total = len(df[col])
        df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=dayfirst)
        convertidas = df[col].notna().sum()
        nulas = df[col].isna().sum()
        print(f"{col}: {convertidas} convertidas | {nulas} no convertibles de {total}")

    return df

def convertir_a_float(df, columna, nombre_df):
    if columna not in df.columns:
        print(f"La columna '{columna}' no existe en {nombre_df}.")
        return df

    total = len(df)

    original = df[columna].copy()

    df[columna] = pd.to_numeric(df[columna], errors="coerce")

    convertidas = df[columna].notna().sum()
    nulas = df[columna].isna().sum()

    print(f"'{columna}' convertida a float: {convertidas} correctas | {nulas} no convertibles de {total}")

    if nulas > 0:
        print(f"Valores no convertibles:")
        print(original[df[columna].isna()].head(10))

    return df


print(f"--------------------------- Ventas ---------------------------")
ventas = convertir_a_float(ventas_limpio, "precio", "Ventas")
ventas = convertir_fechas(ventas_limpio, ["fecha_venta"], "Ventas")

print(f"\n--------------------------- Marketing ---------------------------")
marketing = convertir_fechas(marketing_limpio,["fecha_inicio","fecha_fin"], "Marketing")
```

### 5锔 Transformaci贸n de datos

```python
def detectar_columna_producto(df, patrones=("producto", "id_producto", "sku", "articulo", "art铆culo")):
    for col in df.columns:
        nombre = str(col).lower()
        if any(p in nombre for p in patrones):
            return col
    return None

def agregar_rendimiento_por_producto(df_ventas, col_prod, col_precio="precio", col_cantidad="cantidad"):
    ventas_enriquecidas = df_ventas.assign(ingreso = df_ventas[col_precio] * df_ventas[col_cantidad])
    resumen = (
        ventas_enriquecidas
        .groupby(col_prod, as_index=False)
        .agg(
            ingreso_total = ('ingreso', 'sum'),
            unidades = (col_cantidad, 'sum'),
            precio_promedio = (col_precio, 'mean'),
            registros = ('ingreso', 'size')
        )
    )
    return resumen

def calcular_umbral_percentil(df_resumen, columna_metricas="ingreso_total", q=0.80):
    return df_resumen[columna_metricas].quantile(q)

def filtrar_alto_rendimiento(df_resumen, umbral_p80):
    df_top = (
        df_resumen
        .query("ingreso_total >= @umbral_p80", engine="python")
        .sort_values(by=["ingreso_total", "unidades"], ascending=[False, False], ignore_index=True)
    )
    return df_top


col_producto = detectar_columna_producto(ventas_limpio)
if col_producto is None:
    raise ValueError("No se encontr贸 columna de producto. Renombr谩 una columna a 'producto' o similar.")

tabla_rendimiento = agregar_rendimiento_por_producto(
    ventas_limpio,
    col_prod=col_producto,
    col_precio="precio",
    col_cantidad="cantidad"
)

# Calcular el percentil 80
p80 = calcular_umbral_percentil(tabla_rendimiento, columna_metricas="ingreso_total", q=0.80)

# Filtrar productos de alto rendimiento
ventas_top = filtrar_alto_rendimiento(tabla_rendimiento, p80)

# Mostrar resultados
print(f"Columna de producto detectada: {col_producto}")
print(f"Umbral (percentil 80) de ingreso_total: {p80:,.2f}")
print("\n Productos de alto rendimiento (top 20%):")
display(ventas_top.head(10))
```

### 6锔 Agregaci贸n

```python
def detectar_columna(df, patrones):
    for c in df.columns:
        nombre = str(c).lower()
        if any(p in nombre for p in patrones):
            return c
    return None


col_categoria = detectar_columna(
    ventas_limpio,
    patrones=["categoria", "categor铆a", "categoria_producto", "rubro", "segmento"]
)
if col_categoria is None:
    raise ValueError("No se encontr贸 una columna de categor铆a (p. ej., 'categoria' o 'rubro').")

if "ingreso" not in ventas_limpio.columns:
    base_cat = ventas_limpio.assign(ingreso = ventas_limpio["precio"] * ventas_limpio["cantidad"])
else:
    base_cat = ventas_limpio.copy()

resumen_categoria = (
    base_cat
    .groupby(by=col_categoria, as_index=False, dropna=False)
    .agg(
        ingreso_total = ("ingreso", "sum"),
        unidades = ("cantidad", "sum"),
        ventas = ("ingreso", "size"),
        precio_promedio = ("precio", "mean")
    )
    .sort_values(by="ingreso_total", ascending=False, na_position="last", ignore_index=True)
)

resumen_categoria = resumen_categoria.assign(
    ticket_promedio_por_venta = resumen_categoria["ingreso_total"] / resumen_categoria["ventas"]
)


print(f"Columna de categor铆a detectada: {col_categoria}")
print("Resumen por categor铆a (ordenado por ingreso_total):")
display(resumen_categoria.head(20))
```
